```python
# Group by 'City' and 'Season' and count the number of seasons for each city
city_season_counts = matches.groupby(['city', 'Season']).size().reset_index(name='SeasonsCount')

# Group by 'City' and sum the counts to get the total number of seasons hosted by each city
city_total_seasons = city_season_counts.groupby('city')['SeasonsCount'].sum()

# Find the city with the maximum total seasons
most_seasons_city = city_total_seasons.idxmax()

# Print the result
print(f"City hosting the most number of seasons: {most_seasons_city}")
```

```python
unique_seasons = matches['Season'].unique()
season_counts = matches['Season'].value_counts()

# Sort unique_seasons
sorted_unique_seasons = sorted(unique_seasons)

for season in sorted_unique_seasons:
    count = season_counts[season]
    print(f"Season: {season}, Count: {count}")
```

Which year has the most number of matches played?

Season: IPL-**2013**, Count: 76


```python
# Filter the dataset for Mumbai Indians in 2011
mi_2011_matches = matches[(matches['winner'] == 'Mumbai Indians') & (matches['Season'] == 'IPL-2011')]

# Find the maximum wins for Mumbai Indians in 2011
max_wins = mi_2011_matches['winner'].value_counts().max()

print(f"Maximum wins by Mumbai Indians in 2011: {max_wins}")
```

#Deliveries dataset


```python
deliveries = pd.read_csv("/content/drive/MyDrive/IPL_DATASET/deliveries.csv", encoding = "ISO-8859-1")
```

```python
deliveries.head()
```

```python
deliveries.columns
```

```python
# Group by bowler and calculate the total number of extras
bowler_extras = deliveries.groupby("bowler")["extra_runs"].sum().reset_index()

# Find the bowlers with the maximum number of extras
max_extras = bowler_extras.loc[bowler_extras['extra_runs'].idxmax(), 'extra_runs']
top_bowlers = bowler_extras[bowler_extras['extra_runs'] == max_extras]['bowler'].tolist()

print(f"The bowlers with the maximum number of extras are: {', '.join(top_bowlers)} with {max_extras} extras.")
```

```python
deliveries.tail()
```

```python
deliveries['batsman'].unique()
```

```python
# Filter deliveries for run-outs
runouts = deliveries[deliveries['player_dismissed'] == 'run out']

# Group by batsman and count the number of run-outs
batsman_runouts = runouts.groupby("batsman")["player_dismissed"].sum().reset_index()

# Sort by run-outs in descending order
sorted_batsmen = batsman_runouts.sort_values(by="player_dismissed", ascending=False).reset_index(drop=True)

# Get the top three batsmen with the most run-outs
top_three_batsmen = sorted_batsmen.head(3)

print("Top three batsmen with the most run-outs across seasons:")
print(top_three_batsmen)
```

```python
# Filter deliveries for V. Kohli when the bowler is JJ Bumrah
kohli_bumrah = deliveries[(deliveries['batsman'] == 'V Kohli') & (deliveries['bowler'] == 'JJ Bumrah')]

# Calculate the total runs scored
total_runs_kohli_bumrah = kohli_bumrah['total_runs'].sum()

print(f"Total runs scored by V. Kohli when the bowler was JJ Bumrah: {total_runs_kohli_bumrah}")
```

```python
caught_and_bowled_dismissals = deliveries[deliveries['dismissal_kind'] == 'caught and bowled']

# Group by batsman and count the number of dismissals via caught and bowled
dismissals_count = caught_and_bowled_dismissals['batsman'].value_counts()

# Get the player with the maximum number of dismissals via caught and bowled
max_dismissals_player = dismissals_count.idxmax()
max_dismissals_count = dismissals_count.max()

print(f"The player dismissed the maximum number of times via caught and bowled is {max_dismissals_player} with {max_dismissals_count} dismissals.")
```

```python
#let's visualize match data by win_by_runs using swarmplot

plt.figure(figsize = (20,10))
sns.swarmplot(y="win_by_runs", x ="Season", data = matches,  color="c")
plt.xticks(rotation= 90)
plt.show()
```

#### with this above plot we can interpret most matches have been won in the ranger of 20-60 runs.

```python
#let's visulaize match data by 'toss_decision' using countplot

plt.figure(figsize = (12,6))
sns.countplot(x = "Season", hue = "toss_decision", data = matches)
plt.xticks(rotation = "vertical")
plt.show()
```

### Team's mostly have chosed to bowl when they had won the toss, and in the recent years from 2016 to 2019 choosing bowl has become a default option.

```python
# let's find out the team's which all have won the match who have chosed to bowl first

num_of_wins = (matches.win_by_wickets>0).sum()
num_of_loss = (matches.win_by_wickets==0).sum()
labels = ['Wins', 'Loss']
total = float(num_of_wins + num_of_loss)
sizes = [(num_of_wins/total) * 100, (num_of_loss/total)*100]
colors = ['orange', 'pink']
plt.pie(sizes, labels = labels, colors = colors, autopct = "%1.1f%%", shadow = True, startangle=90)
plt.title("Win percentage batting second")
plt.show()
```

#### as we can see team's which have chosed to bowl first won the match and percentage of winning the match is 53.7%

```python
#let's explore the performance of the batsman and find out who were the best performer

best_p = deliveries.groupby("batsman")['batsman_runs'].agg('sum').reset_index().sort_values(by = 'batsman_runs', ascending = False).reset_index(drop = True)
best_p = best_p.iloc[:10, :]

labels = np.array(best_p['batsman'])
ind = np.arange(len(labels))
width = 0.7
fig, ax = plt.subplots(figsize = (15,8))
rects = ax.bar(ind, np.array(best_p['batsman_runs']), width = width, color = 'blue')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation = "vertical")
ax.set_ylabel("Count")
ax.set_title("Batsman - Top Run Scorer in IPL")
ax.set_xlabel("Batsman's Name")

plt.show()
```

##Something exciting here!!!

```python
#let's see the number of 6's

best_s = deliveries.groupby("batsman")["batsman_runs"].agg(lambda x: (x==6).sum()).reset_index().sort_values(by = "batsman_runs", ascending = False).reset_index(drop = True)
best_s = best_s.iloc[:10, :]
labels = np.array(best_s['batsman'])
ind = np.arange(len(labels))
width = 0.8
fig, ax = plt.subplots(figsize = (15,8))
rects = ax.bar(ind, np.array(best_s['batsman_runs']), width = width, color = "pink")
ax.set_xticks(ind+((width)/2))
ax.set_xticklabels(labels, rotation = 90)
ax.set_ylabel("Count")
ax.set_title("Batsman - Most No. of Sixes in IPL")
ax.set_xlabel("Batsman's Name")

plt.show()
```

## Chris Gayle  has shot more sixes followed by 'AB de Villiers' and 'MS Dhoni'

```python
## let's find the number of 4's

best_f = deliveries.groupby("batsman")["batsman_runs"].agg(lambda x: (x == 4).sum()).reset_index().sort_values(by = "batsman_runs", ascending = False).reset_index(drop = True)
best_f = best_f.iloc[:10, :]
labels = np.array(best_f["batsman"])
ind = np.arange(len(labels))

width = 0.6
fig, ax = plt.subplots(figsize=(15,8))
rects = ax.bar(ind, np.array(best_f['batsman_runs']),  width = width, color = "blue")
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation = 90)
ax.set_ylabel("Count")
ax.set_title("Batsman - Most no. of FOURS in IPL")
ax.set_xlabel("Batsman's Name")

plt.show()
```
